```python
filenames = os.listdir("./Gas")
for filename in filenames:
    if filename.split("_")[1]!='gas':
        namelist=filename.split("_")
        prefix= namelist[0][:-3]
        namelist[0] = prefix
        namelist.append(namelist[1])
        namelist[1] = "gas"
        new_name = '_'.join(namelist)
        print(filename)
        os.rename(f"./Gas/{filename}", f"./Gas/{new_name}")
```
